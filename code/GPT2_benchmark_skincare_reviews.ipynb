# Building the Model
## Initialization
Once we have both frameworks installed, we can import what we need and initialize our tokenizer and model like so:

```python
import torch
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# initialize tokenizer and model from pretrained GPT2 model
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
model = GPT2LMHeadModel.from_pretrained('gpt2')
```

## Tokenization

The tokenizer is used to translate between human-readable text and numeric indices. These indices are then mapped to word embeddings (numerical representations of words) by an embedding layer within the model.

All we need to do to tokenizer our input text is call the tokenizer.encode method like so:

```python
input = tokenizer.encode("This skincare product I bought is great", return_tensors='pt')
```

Because we are using PyTorch, we add return_tensor='pt', if using TensorFlow, we would use return_tensor='tf'

## Generate
Now that we have our tokenization input text, we can begin generating text with GPT-2! All we do is call the model.generate method:

```python
output1 = model.generate(input, max_length=200, do_sample=True)
output2 = model.generate(input, max_length=200, do_sample=True, temperature=.7)
output3 = model.generate(input, max_length=200, do_sample=True, temperature=.7, top_k=50)
```

Here we set the maximum number of tokens to generate as 200. We also add do_sample=True to stop the model from just picking the most likely word at every step.

The top_k and temperature arguments can be used to modify our outputs' coherence/randomness

- temperature runs input into model, so temperature > 1 will create more random token than the default -> more creative but can be weird text if it is too high. Anything < 1 is less random. We can add more randomness with temperature — the default value is 1, a high value like 5 will produce a pretty nonsensical output. Turning the temperature down below 1 will produce more linear but less creative outputs

- Top k limits the sample tokens to the top rated token that the model is predicting so we can put 50 & this will alter our output generally. It limits the sample tokens to a given number of the most probable tokens. This results in text that tends to stick to the same topic (or set of words) for a longer period of time.

## Decoding
Our generate step outputs an array of tokens rather than words. To convert these tokens into words, we need to .decode them. This is easy to do:

All we need to add is skip_special_tokens=True to avoid decoding special tokens that are used by the model, such as the end of sequence token.

At this point, all we need to do is print our output!

```python
text = tokenizer.decode(output1[0], skip_special_token=True)
print(text)
```

```python
text = tokenizer.decode(output2[0], skip_special_token=True)
print(text)
```

```python
text = tokenizer.decode(output3[0], skip_special_token=True)
print(text)
```
