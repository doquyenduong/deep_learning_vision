```python
import torch
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# initialize tokenizer and model from pretrained GPT2 model
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
model = GPT2LMHeadModel.from_pretrained('gpt2')
```

```python
from datasets import load_dataset
dataset = load_dataset('amazon_us_reviews', 'Mobile_Electronics_v1_00', split='train')
```

```python
# convert to pandas
df = pd.DataFrame(dataset)
df.head()
```

```python
input = tokenizer.encode(df.review_body[0], return_tensors='pt')
```

```python
output = model.generate(input, max_length=200, do_sample=True)
```

```python
text = tokenizer.decode(output[0], skip_special_token=True)
print(text)
```
